```python
# Import the needed libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
# Load the dataset as a DataFrame
df = pd.read_csv(r"C:\Users\USER\Downloads\energydata_complete.csv")
df.head()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
X = df["T2"]
y = df["T6"]

# Reshaping the X array
X = np.array(X).reshape(-1, 1)

# Create an instance of the LinearRegression model
model = LinearRegression()

# Fit the linear regression model
model.fit(X, y)

# Predict the target variable using the fitted model
y_pred = model.predict(X)

# Calculate the R^2 score
r2 = r2_score(y, y_pred)

# Print the R^2 score rounded to 2 decimal places
print(f"R-squared: {r2:.2f}")
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error


# Remove columns "date" and "lights" from the dataset
columns_to_drop = []
df = df.drop(columns_to_drop, axis=1)

# Separate the features (X) and the target variable (y)
X = df.drop("Appliances", axis=1)
y = df["Appliances"]

# Normalize the features using MinMaxScaler
scaler = MinMaxScaler()
X_normalized = scaler.fit_transform(X)

# Perform a train-test split with a random state of 42
X_train, X_test, y_train, y_test = train_test_split(X_normalized, y, test_size=0.3, random_state=42)

# Fit the multiple linear regression model on the training set
model = LinearRegression()
model.fit(X_train, y_train)

# Predict the target variable for the test set
y_pred = model.predict(X_test)

# Calculate the Mean Absolute Error (MAE)
mae = mean_absolute_error(y_test, y_pred)

# Print the MAE rounded to 2 decimal places
print(f"Mean Absolute Error: {mae:.2f}")
```

```python
# Calculate the Residual Sum of Squares (RSS)
rss = ((y_test - y_pred) ** 2).sum()

# Print the RSS rounded to 2 decimal places
print(f"Residual Sum of Squares: {rss:.2f}")
```

```python
from sklearn.metrics import mean_squared_error

# Calculate the Mean Squared Error (MSE)
mse = mean_squared_error(y_test, y_pred)

# Calculate the Root Mean Squared Error (RMSE)
rmse = np.sqrt(mse)

# Print the RMSE rounded to 3 decimal places
print(f"Root Mean Squared Error: {rmse:.3f}")
```

```python
from sklearn.metrics import r2_score

# Calculate the coefficient of determination (R-squared)
r2 = r2_score(y_test, y_pred)

# Print the R-squared rounded to 2 decimal places
print(f"Coefficient of Determination (R-squared): {r2:.2f}")
```

```python
from sklearn.linear_model import Lasso

# Train a Lasso regression model with alpha=0.001
lasso_model = Lasso(alpha=0.001)
lasso_model.fit(X_train, y_train)

# Obtain the feature weights from the Lasso model
feature_weights = lasso_model.coef_

# Count the number of features with non-zero feature weights
non_zero_features = sum(feature_weights != 0)

# Print the number of features with non-zero feature weights
print(f"Number of features with non-zero weights: {non_zero_features}")
```
